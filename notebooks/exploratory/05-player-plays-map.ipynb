Let's try to load it now

```python
loaded_game = {}
with open("game.txt", "r") as fp:
    # Load the dictionary from the file
    loaded_game = json.load(fp)
print(json.dumps(loaded_game, indent=4))
```

Now I can translate it back to its original format

```python
from src.game_generation.utils.data_structures import Pos

for i in loaded_game["trajectory"]:
    print(Pos(*list(i.values())))
```
